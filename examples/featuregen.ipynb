# `FeatureGenerator` Example

This notebook demonstrates how to use the `FeatureGenerator` transformer to automatically create polynomial and interaction features from your dataset.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
from transfory.featuregen import FeatureGenerator
```

## 1. Create Sample Data

First, let's create a simple DataFrame with a few numeric columns.

```python
df = pd.DataFrame({
    "Height": [150, 160, 170, 180],
    "Weight": [50, 60, 70, 80]
})

print("Original DataFrame:")
df
```

## 2. Apply the FeatureGenerator

Now, we'll instantiate `FeatureGenerator`. By default, it creates polynomial features up to `degree=2` and includes interaction terms (`include_interactions=True`).

```python
# Create an instance of the FeatureGenerator
fg = FeatureGenerator(degree=4)

# Fit to the data and transform it
new_df = fg.fit_transform(df)

print("DataFrame with New Features:")
new_df
```
